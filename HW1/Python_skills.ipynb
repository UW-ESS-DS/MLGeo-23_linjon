```python
########
# ARRAYS

# make a 1D array of Fibonacci numbers

# Create an empty array
Fibonacci = np.zeros(500)

# Fill the beginning of the sequence manually because the fibonacci algorith requires a starting number
Fibonacci[0] = 0
Fibonacci[1] = 1
# Calculate each next fibonacci number and add it to the end of the array
for k in range(len(Fibonacci)-2):
    Fibonacci[k+2] = Fibonacci[k+1] + Fibonacci[k]

#print(Fibonacci)
```

```python
########
# Plotting some stuff and declaring some functions

# A bumpy surface

# Define some surface as a function of two variables:
bumps = lambda x, y: np.sin(x) + np.sin(y)


# First create x and y spaces
x = np.linspace(-7, 7, 40)
y = np.linspace(-7, 7, 40) 


# Then create a mesh with equally spaced points
# in the x and y directions.
X, Y = np.meshgrid(x, y)


# Setup the figure
fig = plt.figure() # Create a figure
ax = plt.axes(projection='3d') # Make it a "3D" figure
surface = ax.plot_surface(X, Y, bumps(X,Y), cmap = 'viridis') 
fig.colorbar(surface)
ax.view_init(30,35)
ax.set_zlim(0,4)
ax.set_xlabel('$x$', fontsize=10)
ax.set_ylabel('$y$', fontsize=10) 
ax.set_zlabel('$f(x, y)$', fontsize=10, rotation=90) 
ax.set_title("Plot of a bumpy surface")

# Aaand save the plot as a .png image file:
plt.savefig('bumps.png', bbox_inches='tight')

```

```python


####Error function curve fitting

#Generate an array of random numbers
rng = np.random.default_rng()
rand_array = rng.random((80,2))

x = rand_array[:,0]
y = rand_array[:,1]

f = lambda x: 4 / 5 *np.sqrt(1 - x**2) + np.sqrt(np.abs(x))
xspace = np.linspace(-1, 1, 1000)
plt.plot(x, y, 'ro', markersize = 3, label = "Meaningless red points")
plt.plot(xspace, f(xspace), 'bo', markersize = 2, label = "Meaningless blue points")
plt.title("Two sets of completely meaningless points")
plt.legend()
plt.plot()
```
